```python
import numpy as np
import pandas as pd
import time
import numpy.linalg as ln
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
with open('1.txt') as f:
    contents = f.read()


b = [0]*60
a = [0]*60
c = contents.split("\n")
#print(len(contents))
for i in range(60):
    
    b[i] = c[i].split(" ")
    b[i][0] = float(b[i][0])
    b[i][1] = float(b[i][1])
    a[i] = b[i][1]


x = np.arange(5, 305, 5)


plt.plot(x, a, 'o')
a_no = a
plt.ylabel('time, sec')
plt.xlabel('matrix size')
plt.title("Время умножения матриц от размера матриц для наивной однопоточной реализации") # заголовок
plt.show()
```
